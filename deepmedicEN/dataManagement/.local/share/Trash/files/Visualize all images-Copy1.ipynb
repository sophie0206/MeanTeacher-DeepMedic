```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
brainmask = nib.load('brainmask.nii.gz').get_data()
Flair = nib.load('Flair_subtrMeanDivStd.nii.gz').get_data()
T1c = nib.load('T1c_subtrMeanDivStd.nii.gz').get_data()
OTMultiClass = nib.load('OTMultiClass.nii.gz').get_data()
weightMap0 = nib.load('weightMap0.nii.gz').get_data()
weightMap1 = nib.load('weightMap1.nii.gz').get_data()
```

```python
def show_slices(img):
    #img = np.transpose(img, [2,1,0])
    w, h, l = img.shape
    slices = [img[w//2,:,:], img[:,h//2,:], img[:,:,l//2]]
    fig, axes = plt.subplots(1, len(slices))
    for i, slice in enumerate(slices):
        axes[i].imshow(slice.T, cmap='gray')
```

```python
show_slices(brainmask)
show_slices(Flair)
show_slices(T1c)
show_slices(OTMultiClass)
show_slices(weightMap0)
show_slices(weightMap1)
```
