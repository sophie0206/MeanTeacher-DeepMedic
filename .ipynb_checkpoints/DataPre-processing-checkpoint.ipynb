```python
def show_slices(img):
    #img = np.transpose(img, [2,1,0])
    w, h, l = img.shape
    
    slices = [img[w//2,:,:], img[:,h//2,:], img[:,:,l//2]]
    fig, axes = plt.subplots(1, len(slices))
    for i, slice in enumerate(slices):
        axes[i].imshow(slice,cmap='gray',origin='lower')
```

```python
def set_params(filename):
    sigma, I0 = 0, 0
    if 't1ce' in filename:
        print('t1ce')
        sigma=31 
        I0=99
    if 't2' in filename:
        print('t2')
        sigma=37
        I0=55
    if 'flair' in filename:
        print('flair')
        sigma=30
        I0=75
    return sigma, I0
```

```python
def zero1_normalize(volume):
    mean = volume[volume != volume[0, 0, 0]].mean()
    std  = volume[volume != volume[0, 0, 0]].std()
    out = (volume - mean)/std
    return out
```

```python
def to_range(data, upper_bound, lower_bound):
    d1 = (data-data.min()) / (data.max()-data.min())
    rage = upper_bound - lower_bound
    normalized = d1 * rage + lower_bound
    return normalized
```

```python
def robust_normalization(data, sigma, I0, upper_bound, lower_bound):
    normalized = to_range(data, 255, 0)
    
    w,h,l = data.shape
    re = data
    for i in range(w):
        for j in range(h):
            for k in range(l):
                if data[i, j, k] > 0:
                    re[i,j,k] = normalized[i,j,k] * sigma + I0
    out = to_range(re, 255, 0)
    return re
```

```python
def pre_process(idata, filename):
    sigma, I0 = set_params(filename)
    print(sigma, I0)
    normalized = idata
    if((sigma>0) & (I0>0)):
        normalized = robust_normalization(idata, sigma, I0, 0, 255)
    out = zero1_normalize(normalized)
    return out
```

```python
def save_img(idata, filepath, affine):
    transposed = idata.astype('float32')
    final_img = nib.Nifti1Image(transposed, affine)
    nib.save(final_img, filepath)
```

```python
def load_data(path, path2Seg):
    annotated = os.listdir(path)
    
    segs = os.listdir(path2Seg)
    
    for m in annotated:
        img = nib.load(path+'/'+ m + '/norm_result.nii.gz')
        affine = img.affine
        idata = img.get_fdata()
        #print(path2Seg + '/' + seg)
        img_seg = nib.load(path2Seg + '/' + m + '.nii.gz')
        affine_seg = img_seg.affine
        seg_data = img_seg.get_data()
        seg_data[seg_data > 0] = 1
        show_slices(img_seg.get_fdata())
        
        if not os.path.exists('/data/wenhuicui/stroke/labeled_train/' + m):
            os.makedirs( '/data/wenhuicui/stroke/labeled_train/' + m)
        
        processed = zero1_normalize(idata)
        show_slices(processed)
        #print(processed[100, 100, 15])
        save_img(processed,  '/data/wenhuicui/stroke/labeled_train/' + m + '/img.nii.gz', affine)
        save_img(seg_data, '/data/wenhuicui/stroke/labeled_train/' + m + '/seg.nii.gz', affine_seg)
        save_img(seg_data, '/data/wenhuicui/stroke/labeled_train/'+ m + '/weightMap1.nii.gz', affine_seg)
        
        
        idata[idata == idata[0, 0, 0]] = 0
        idata[idata != idata[0, 0, 0]] = 1
        brainmask = idata
        show_slices(brainmask)
        save_img(brainmask, '/data/wenhuicui/stroke/labeled_train/'+ m + '/brainmask.nii.gz', affine)
        
        seg = seg_data
        wmp0 = brainmask
        wmp0[seg > 0] = brainmask[0, 0, 0]
        show_slices(wmp0)
        save_img(wmp0, '/data/wenhuicui/stroke/labeled_train/'+ m + '/weightMap0.nii.gz', affine)
```

```python
def load_data_unlabeled(path, labeled, path2Seg):
    allData = os.listdir(path)
    
    for m in allData:
        if m in labeled:
            continue
            
        img = nib.load(path+'/'+ m + '/norm_result.nii.gz')
        affine = img.affine
        idata = img.get_fdata()
        #print(path2Seg + '/' + seg)
        img_seg = nib.load(path2Seg + '/pred_' + m + '_Segm.nii.gz')
        affine_seg = img_seg.affine
        seg = img_seg.get_data()
        seg[seg > 1] = 1
        seg[seg < 1] = 0
    
        #show_slices(img_seg.get_fdata())
        
        if not os.path.exists('/data/wenhuicui/stroke/unlabeled_train/' + m):
            os.makedirs( '/data/wenhuicui/stroke/unlabeled_train/' + m)
        
        processed = zero1_normalize(idata)
        save_img(processed,  '/data/wenhuicui/stroke/unlabeled_train/' + m + '/img.nii.gz', affine)
        save_img(seg, '/data/wenhuicui/stroke/unlabeled_train/' + m + '/seg.nii.gz', affine_seg)
        save_img(seg, '/data/wenhuicui/stroke/unlabeled_train/'+ m + '/weightMap0.nii.gz', affine_seg)
        
        idata[idata == idata[0, 0, 0]] = 0
        idata[idata != idata[0, 0, 0]] = 1
        brainmask = idata
        show_slices(brainmask)
        save_img(brainmask, '/data/wenhuicui/stroke/unlabeled_train/'+ m + '/brainmask.nii.gz', affine)
        
        
        wmp0 = brainmask
        wmp0[seg > 0] = brainmask[0, 0, 0]
        #show_slices(wmp0)
        save_img(wmp0, '/data/wenhuicui/stroke/unlabeled_train/'+ m + '/weightMap1.nii.gz', affine)
```

```python
def makeTDGtLabels(path):
    files = os.listdir(path)
    for f in files:
        if f == '.DS_Store':
            continue
        mods = os.listdir(path+'/'+f)
        for m in mods:
            if 'brainmask' in m:
                image = nib.load(path+'/'+f+'/'+m)
                affine = image.affine
                idata = image.get_data()

                idata[idata > 0] = 0
                show_slices(idata)

                save_img(idata, path+'/'+f+'/'+'TDlables.nii.gz', affine)
```
